```python
#Print in a cleaner formatted line (f-string)
def result(hours_studied, attendance):
    return (hours_studied >= 4) and (attendance > 75)

for s in students:
    s["passed"] = result(s["hours_studied"], s["attendance"])
    print(f"Name={s['name']}, Hours={s['hours_studied']}, Attendance={s['attendance']}, Passed={s['passed']}")
```
